```python
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
data_dict = pickle.load(open('./data.pickle', 'rb'))
```

```python
data_dict
```

```python
data = data_dict['data']
labels = data_dict['labels']
```

### Determine the maximum length among the elements in 'data'

```python
max_length = max(len(element) for element in data)
```

```python
max_length
```

### Pad the elements in 'data' with zeros to make them have the same length

```python
padded_data = [element + [0] * (max_length - len(element)) for element in data]
```

### Convert 'padded_data' and 'labels' to NumPy arrays

```python
data = np.asarray(padded_data)
labels = np.asarray(labels)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, shuffle=True, stratify=labels)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
score = accuracy_score(y_predict, y_test)
```

```python
print('{}% of samples were classified correctly !'.format(score * 100))
```

```python
f = open('model.p', 'wb')
pickle.dump({'model': model}, f)
f.close()
```
